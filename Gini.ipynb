# Análisis de Desigualdad de Ingresos con el Índice de Gini
Este proyecto calcula y visualiza la distribución del ingreso en un grupo de personas del grupo 601 de la Facultad de Economía de la UNAM, utilizando la **curva de Lorenz** y el **índice de Gini**.  

Herramientas utilizadas:  
- Python
- NumPy  
- Matplotlib

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
import statistics as  statistics
```

## Datos
Los datos representan la **frecuencia de gasto semanal** en distintos intervalos del grupo.

## Cálculo de frecuencias
Se calculan las frecuencias acumuladas y relativas.

```python
#1 CALCULAMOS LA FRECUENCIA
print("FRECUENCIA")
f2=(4,13,6,4,2,1)
f1=(4,13,6,4,2,1)
f=np.array(f1)
n=sum(f)
print(f1)
print(n)
```

```python
#2 DETERMINAMOS LOS INTERVALOS
print("INTERVALO")
lim_inf=(0,400,800,1200,2000,2400)
lim_sup=(400,800,1200,1600,2400,2500)
```

```python
#3 CALCULAMOS LA FRECUENCIA ACUMULADA
print("FRECUENCIA ACUMULADA")
fa=np.cumsum(f)
print(fa)

#4 CALCULAMOS LA FRECUENCIA RELATIVA
print("FRECUENCIA RELATIVA")
fr=(f/n)
print(fr)
fr1=(fr*100)
fr1=np.round(fr1,2)
print(fr1)

#5 CALCULAMOS LA FRECUENCIA RELATIVA ACUMULADA
print("FRECUENCIA RELATIVA ACUMULADA")
fra=fa/n
print(fra)
fra1=(fra*100)
xi=np.round(fra1,2)
print("x1")
print(xi)

#6 CALCULAMOS LA FRECUENCIA DE LA MARCA DE CLASE
print("Mc")
mc=[200,600,1000,1400,1800,2600]
print(mc)
fmc=mc*f
print(fmc)

#7 CALCULAMOS LA SUMA DE LA MARCA DE CLASE
print("TOTAL DE FMC")
tfmc=sum(fmc)
print(tfmc)

#8 CALCULAMOS LA FRECUENCIA ACUMULADA DE LA MARCA DE CLASE
print("Fmc ACUMULADA")
fmca=np.cumsum(fmc)
print(fmca)

#9 CALCULAMOS LA PROBABILIDAD DE LA FRECUENCIA ACUMULADA DE LA MARCA DE CLASE
print("Prob. de Fmc. ACUMULADA")
pfmac=fmca/tfmc
print(pfmac)
pfmac1=pfmac*100

#10 CALCULAMOS LA COLUMNA DE yi
print("yi")
yi=np.round(pfmac1,0)
print(yi)

#11 CALCULAMOS LA MEDIA DE LOS DATOS OBTENIDOS
print("Media del gasto semanal del grupo")
media=tfmc/n
print(media)

#12 CALCULAMOS LA FRECUENCIA POR DESVIACION ABSOLUTA
print ("Frecuencia por desviación absoluta")
MC = np.array(mc)
fmcx=f1*(abs(MC-880))
print(fmcx)

#13 CALCULAMOS LA COLUMNA QUE MULTIPLICA LA COLUMNA Xi POR Yi
print ("Xi(Yi+1)")
matriz1 = np.array([13,56,76,90,96])
matriz2 = np.array([32,55,76,90,100])
XiYi=matriz1*matriz2
print(XiYi)

#14 SE SUMA LOS RESULTADOS DE Xi(Yi+1)
print ("Sumatoria de Xi(Yi+1)")
sumXiYi=sum(XiYi)
print(sumXiYi)


#15 CALCULAMOS LA COLUMNA QUE MULTIPLICA LA COLUMNA Yi POR Xi
print ("(Xi+1)Yi")
matriz3 = np.array([3,32,55,76,90])
matriz4 = np.array([56,76,90,96,100])
YiXi=matriz3*matriz4
print(YiXi)

#16 SE SUMA LOS RESULTADOS DE Xi(Yi+1)
print ("Sumatoria de (Xi+1)Yi")
sumYiXi=sum(YiXi)
print(sumYiXi)
```

## Cálculo del Índice de Gini
El índice de Gini mide la desigualdad.  
Se obtiene a partir de la diferencia entre el área bajo la línea de igualdad perfecta y la curva de Lorenz.


```python
#17 CON LOS DATOS ANTERIORMENTE OBTENIDOS CALCULAMOS EL INDICE DE GINI
print ("Indice de GINI del grupo")
Gini=(sumXiYi-sumYiXi)/(100**2)
print(Gini)
```

## Curva de Lorenz
La curva de Lorenz compara la **distribución acumulada de la población** contra la **distribución acumulada del ingreso**.


```python
#18 GRAFICAMOS LA CURVA DE LORENZ DE LA DIFERENCIA EN EL INGRESO SEMANAL
print ("CURVA DE LORENZ, GASTO SEMANAL DEL GRUPO")
x=(xi)
y=(yi)
plt.plot( x, x, marker='o')
plt.plot(x, y, marker='o')
plt.xlabel(' % POBLACIÓN')
plt.ylabel(' % GASTO')
plt.legend(["LINEA DE IGUALDAD", "CURVA DE LORENZ"], bbox_to_anchor=(0.75, -0.15), ncol=2)
plt.title('CURVA DE LORENZ DEL GASTO SEMANAL EN EL GRUPO')
plt.grid()
plt.show()
```

## Histograma de frecuencias
Distribución de gasto semanal del grupo en intervalos.


```python
#19 GRAFICAMOS EL HISTOGRAMA DE FRECUENCIAS DEL GASTO
print ("HISTOGRAMA DE FRECUENCIAS, GASTO SEMANAL DEL GRUPO")
marca_clase = [0, 401, 801, 1201, 1601, 2401]
frecuencias = [4, 13, 6, 4, 2, 1]
ancho_clase = 400
plt.bar(marca_clase, frecuencias, width=ancho_clase, align='center')
plt.xlabel('INTERVALOS')
plt.ylabel('FRECUENCIA')
plt.title('HISTOGRAMA')
plt.grid(True)
plt.show()
```

## Conclusión
El índice de Gini calculado para el grupo fue **0.23**, lo que refleja una distribución de gasto relativamente equitativa.  

Este tipo de análisis es aplicable en estudios de desigualdad económica y permite visualizar de forma clara la dispersión de ingresos en una población.
